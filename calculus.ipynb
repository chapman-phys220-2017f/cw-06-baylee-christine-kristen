## First Derivative as a Matrix

The range of a function is represetned as an $n$ dimension matrix:
$$f = \left( \begin{array}{ccc}
f_0 \\
f_1 \\
... \\
f_{n-1}\end{array} \right) $$
while the derivative is represented as an $n$ x $n$ matrix:
$$D=  \left( \begin{array}{ccc}
a & b & c \\
d & e & f \\
g & h & i \end{array} \right) .$$
If the derivative is known using the forward difference, central difference, and backwards difference formulas, one can begin to think of values to fill in the $D$ matrix. The forwards difference formula shows that the derivative at $f_0$ must be $\frac{f_1- f_0}{dx}$; the backwards difference shows that the derivative at $f_{n-1}$ must be $\frac{f_{n-1} - f_{n-2}}{dx}$; the central difference formula shows that the derivative at an interior point $f_k$ must be $\frac{f_{k+1} - f_{k-1}}{2dx}$. If we turn this into an $n$ dimension matrix like the range, it looks like this:
$$D f = \left( \begin{array}{ccc}
\frac{f_1- f_0}{dx} \\
...\\
\frac{f_{k+1} - f_{k-1}}{2dx} \\
... \\
\frac{f_{n-1} - f_{n-2}}{dx}\end{array} \right). $$
If $D f$ is the product of the two matrices, $f$ and $D$, then one can determine the values of the points within the $D$ matrix.
For a 5x5 matrix, $f$ is $[f_0,f_1,f_2,f_3,f_4]$. If these values are plugged into the previous formulas, points within the $D$ matrix will be filled. The matrix becomes:
$$D=  \left( \begin{array}{ccc}
-1&  1 &  0 &  0 &  0 \\
-0.5 &  0 &  0.5 &  0 & 0\\
0 & -0.5 &  0 &  0.5 &  0\\
0 & 0 & -0.5 &  0 &  0.5\\
0 & 0 & 0 & -1 & 1\end{array} \right) .$$

## Second Derivative as a Matrix

Using the same logic to find the $D$ matrix, the $D2$ matrix is very formulaic. For each $f_k$, where k = [0,n-1], in the $Df$ matrix, the forward, central, or backwards difference formula is applied in order to obtain the second derivative. This yields a $D2f$ matrix such as:
$$D2f = \left( \begin{array}{ccc}
\frac{\frac{f_2- f_0}{2dx}-\frac{f_1-f_0}{dx}}{dx} \\
\frac{\frac{f_3- f_1}{2dx}-\frac{f_1-f_0}{dx}}{2dx} \\
...\\
\frac{\frac{f_{k+2} - f_k}{2dx}-\frac{f_k - f_{k-2}}{2dx}}{2dx} \\
... \\
\frac{\frac{f_{n-2} - f_{n-2}}{2dx}-\frac{f_{n-2}-f_{n-4}}{2dx}}{2dx} \\
\frac{\frac{f_{n-1} - f_{n-2}}{dx}- \frac{f_{n-1}-f_{n-3}}{2dx}}{dx} \end{array} \right). $$
This time, there are still special cases for the endpoints, but also the points right next to the endpoints. The matrix needed to generate the $D2f$ matrix is:
$$D2=  \left( \begin{array}{ccc}
2 & -4 & 2 & 0 & 0 \\
2 & 3  & 0 & 1 & 0 \\
1 & 0 & -2 & 0 & 1 \\
0 & 1  & 0 & 3 & 2\\
0 & 0 & 2 & -4 & 2\end{array} \right) .$$
